# Generating Figures for Genotype-Specific Simulation Runs with ABM

## Grab local data files and configrure plot params

```python
import pandas as pd
import matplotlib.pyplot as plt

apoe33_df = pd.read_csv("..\\data\\APOE_33.csv")
apoe34_df = pd.read_csv("..\\data\\APOE_34.csv")
apoe44_df = pd.read_csv("..\\data\\APOE_44.csv")

plt.rcParams.update({
    "font.size": 14,
    "axes.titlesize": 18,
    "axes.labelsize": 16,
    "legend.fontsize": 13,
    "figure.figsize": (6, 5)
})
```

## Function creating figures for inflammation over time colored by LD accumulated 

```python
def plot_time_vs_inflammation_colorLD(df):

    required = [
        "total_ld_microglia",
        "total_pro_inflammation",
        "step"
    ]

    if not all(c in df.columns for c in required):
        raise ValueError("CSV missing required columns.")

    plt.figure()

    scatter = plt.scatter(
        df["step"],
        df["total_pro_inflammation"],
        c=df["total_ld_microglia"],
        cmap="plasma",
        s=50
    )
    plt.ylim(0, 520)
    plt.xlabel("Time (step)")
    plt.ylabel("Total pro-inflammation")
    plt.title("Time Step vs Pro-Inflammation")

    # Keep color bar scale same (between 1-200)
    cbar = plt.colorbar(scatter)
    plt.clim(1, 200)
    cbar.set_label("Total lipid droplets in microglia")

    plt.tight_layout()
    plt.savefig("time_vs_inflammation_colorLD.pdf")  # VECTOR PDF
    plt.close()

    print("Saved time_vs_inflammation_colorLD.pdf")
```

## Function creating figures for LDAM, DAM, and dead neurons over time 

```python
def plot_percent_neuron_loss_and_ldam_dam(df):
    required = ["step", "neurons_total", "LDAM_fraction", "DAM_fraction",]

    if not all(c in df.columns for c in required):
        raise ValueError("CSV missing required columns.")

    # Initial neuron count at first recorded step
    initial_step = df["step"].min()
    N0 = df.loc[df["step"] == initial_step, "neurons_total"].iloc[0]

    if N0 == 0:
        raise ValueError("Initial neuron count is zero.")

    # Compute percent neuron loss
    df = df.copy()
    df["pct_neuron_loss"] = 100.0 * (1.0 - df["neurons_total"] / N0)
    df["LDAM_fraction"] = 100.0 * df["LDAM_fraction"]
    df["DAM_fraction"] = 100.0 * df["DAM_fraction"] 

    plt.figure(figsize=(6, 5))

    plt.plot(
        df["step"],
        df["pct_neuron_loss"],
        label="Percent Neuron Loss",
        linewidth=2
    )
    plt.plot(df["step"], df["LDAM_fraction"], label="LDAM Fraction")
    plt.plot(df["step"], df["DAM_fraction"], label="DAM Fraction")

    plt.xlabel("Step")
    plt.ylabel("Percent (%)")
    plt.title("Neuron Loss & Microglia States")
    plt.ylim(0, 100)
    plt.legend()

    plt.tight_layout()
    plt.savefig("percent_neuron_loss.pdf")  # vector PDF
    plt.close()

    print("Saved percent_neuron_loss.pdf")
    
```

## Run plot_time_vs_inflammation_colorLD for all three datasets

```python
plot_time_vs_inflammation_colorLD(apoe33_df)
plot_time_vs_inflammation_colorLD(apoe34_df)
plot_time_vs_inflammation_colorLD(apoe44_df)
```

## Run plot_percent_neuron_loss_and_ldam_dam for all APOE 3/3 and APOE 4/4 cases

```python
plot_percent_neuron_loss_and_ldam_dam(apoe33_df)
plot_percent_neuron_loss_and_ldam_dam(apoe44_df)
```
